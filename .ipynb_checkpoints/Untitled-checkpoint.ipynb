### Second exercise: top 10 arrival airports in the world in 2013 

```python
# I have statically set columns indexes of arr_port and pax 
def trimText(obj):
    for idx, val in enumerate(obj):
        obj[idx] = str(val).strip()
    return obj

dfBookings = bookingFile.map(lambda l: l.split("^")).map(trimText)
header = dfBookings.first() #extract header
dfBookings = dfBookings.filter(lambda x :  x != header and len(x) == 38).map(lambda x : (str(x[12]), int(x[34])))
```

```python
def sumBookings(myList):
    ssum = 0
    for idx, val in enumerate(myList):
        ssum = ssum + val[1]
    return ssum

bookingsAiportGrouped = dfBookings.groupBy(lambda tup: tup[0]).map(lambda x: (x[0], sumBookings(list(x[1]))))
bookingsAiportGrouped = bookingsAiportGrouped.sortBy(lambda x : -x[1])
```

```python
# 10 top arrival Airports
bookingsAiportGrouped.take(10)
```

```python
# second approach with spark DataFrame
sparkBookingsDF = dfBookings.toDF(["arr_port", "pax"])
sparkBookingsGroupedDF = sparkBookingsDF.groupBy("arr_port").agg(sum('pax').alias('arr_count')).sort(desc("arr_count"))
```

##### Bonus point: Solve this problem using pandas (instead of any other approach)

```python
def computeTopAirportsDf():
    chunks = pd.read_csv(bFile, sep="^", chunksize=1000000, usecols=[12, 34])
    bookinsPd = [chk.groupby('arr_port', sort=False)['pax'].sum().reset_index() for chk in chunks]
    return pd.concat(bookinsPd).groupby("arr_port").agg({'pax':'sum'}).reset_index().sort_values('pax', ascending=False)
```

```python
topAirports = computeTopAirportsDf()
topAirports.head(10)
```

##### Bonus point: Get the name of the city or airport corresponding to that airport (GeoBases Library is installed)

```python
from GeoBases import GeoBase

def getTopAirportsDf(n=10):
    topAirportsDetail = topAirports.iloc[:n]
    topAirportsDetail = topAirportsDetail.assign(city_name = None, airport_name = None)

    geo_o = GeoBase(data='ori_por', verbose=False)
    for index, row in topAirportsDetail.iterrows(): 
        geoMap = geo_o.get(row['arr_port'].strip(), None)
        topAirportsDetail.loc[index, 'arr_port'] = row['arr_port'].strip()
        topAirportsDetail.loc[index, 'city_name'] = geoMap['city_name_utf']
        topAirportsDetail.loc[index, 'airport_name'] = geoMap['name']
    return topAirportsDetail
```

```python
getTopAirportsDf(10)
```

```python
dfSearches = searchesFile.map(lambda l: l.split("^"))
header = dfSearches.first() #extract header
# in some rows, the data is not complete, so I am removing them for now
dfSearchesPipeLine = dfSearches.filter(lambda x :  x != header and len(x) == 45).filter(lambda x : str(x[6]) in aiportss).map(lambda x : ( dt.strptime(str(x[0]), '%Y-%m-%d') , str(x[6]).strip()))
```

```python
searchedGroupedRDD = dfSearchesPipeLine.groupBy(lambda x : (x[0].year, x[0].month, x[1])).map(lambda x : (dt.strptime(str(x[0][0]) + "-" + str(x[0][1]) + "-01", "%Y-%m-%d"), x[0][2], len(list(x[1]))))
```

```python
searchesFinalDF = searchedGroupedRDD.toDF(["Date", "Destination", "Montly_search"])
```

```python
searchesFinalDF.show(5)
```

```python
import matplotlib.pyplot as plt
def obtainMonthlySearchPlt(dff):
    fig, ax = plt.subplots()
    labels = []

    for key, grp in dff.groupby(['Destination']):
        ax = grp.plot(ax=ax, kind='line', x='Date', y='Montly_search')
        labels.append(key)
    lines, _ = ax.get_legend_handles_labels()
    ax.legend(lines, labels, loc='best')
    plt.title('# of Airport Searches Monthly')
    return plt
```

```python
obtainMonthlySearchPlt(searchesFinalDF.toPandas()).show()
```

##### Bonus point: Solving this problem using pandas (instead of any other approach)

```python
searchesDf = []
for chunk in pd.read_csv(sFile, sep="^", chunksize=1000000, usecols=[0, 6]): # 1000k read from file per chunk
    chunk['Date'] = pd.to_datetime(chunk['Date'], errors='coerce')
    searchesChunk = chunk.loc[chunk['Destination'].isin(aiportss)].groupby([chunk['Date'].dt.to_period("M"), 'Destination'], sort=False).size().reset_index(name='Montly_search')
    searchesDf.append(searchesChunk)
```

```python
monthlyGroupedSearches = pd.concat(searchesDf).groupby(["Date", "Destination"]).agg({'Montly_search':'sum'}).reset_index()
```

```python
monthlyGroupedSearches[:5]
```

```python
obtainMonthlySearchPlt(monthlyGroupedSearches).show()
```

### Bonus exercise: write a Web Service

```python
import json
import cherrypy as appserver

class AirportsController():
    @appserver.expose
    @appserver.tools.json_out()
    def getTopAirports(self, n=10):
        data = {}
        data['status'] = 'success'
        data['key'] = 'value'
        data['n'] = int(n)
        data['topairports'] = json.loads(getTopAirportsDf(int(n)).reset_index().to_json(orient='records'))
        json_data = json.dumps(data)
 
        return json.loads(json_data)
    
    def showOnErr(status, message, traceback, version):
        data = {}
        data['status'] = 'error'
        data['status_code'] = status
        data['err_msg'] = message
        data['msg'] = "go to http://127.0.0.1:9000/getTopAirports?n=10"
        return json.dumps(data)

def startServer():
    appserver.tree.mount(AirportsController(), '/')
    appserver.config.update({'error_page.404': AirportsController.showOnErr})
    appserver.config.update({'error_page.500': AirportsController.showOnErr})
    appserver.config.update({'server.socket_port': 9000})
    appserver.engine.start()
startServer()
```
