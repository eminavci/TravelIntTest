### Second exercise: top 10 arrival airports in the world in 2013 

```python
# I have statically set columns indexes of arr_port and pax 
def trimText(obj):
    for idx, val in enumerate(obj):
        obj[idx] = str(val).strip()
    return obj

dfBookings = bookingFile.map(lambda l: l.split("^")).map(trimText)
header = dfBookings.first() #extract header
dfBookings = dfBookings.filter(lambda x :  x != header).map(lambda x : (str(x[12]), int(x[34])))
```

```python
dfBookings.take(10)
```

```python
def sumBookings(myList):
    ssum = 0
    for idx, val in enumerate(myList):
        ssum = ssum + val[1]
    return ssum

bookingsAiportGrouped = dfBookings.groupBy(lambda tup: tup[0]).map(lambda x: (x[0], len(list(x[1]))))
```

##### Bonus point: Solve this problem using pandas (instead of any other approach)

```python
chunks = pd.read_csv(bFile, sep="^", chunksize=1000000, usecols=[12, 34])
pieces = [chk.groupby('arr_port', sort=False)['pax'].agg(['sum']) for chk in chunks]
agg = pd.concat(pieces).groupby(level=0).sum()
```
